##  Ranking Top 10 Municípos com mais Funding de cada Região por ano 

```python
import pandas as pd
import matplotlib.pyplot as plt
from prettytable import PrettyTable  # Importe a classe PrettyTable da biblioteca prettytable

def atualizar_tabela_e_grafico(criterio):
    # Modifique a linha de agregação para incluir o critério fornecido pelo usuário
    ranking_municipios_ano_regiao = df.groupby(['Ano', 'Região', 'Município'])[criterio].sum().reset_index()

    anos_regioes_unicos = ranking_municipios_ano_regiao[['Ano', 'Região']].drop_duplicates()

    for _, row in anos_regioes_unicos.iterrows():
        ano, regiao = row['Ano'], row['Região']
        print(f"ANO   : {ano}")
        print(f"Região: {regiao}")

        ranking_ano_regiao = ranking_municipios_ano_regiao[(ranking_municipios_ano_regiao['Ano'] == ano) & (ranking_municipios_ano_regiao['Região'] == regiao)]

        ranking_municipio_regiao_ano = ranking_ano_regiao.groupby(['Município'])[criterio].sum().reset_index()

        ranking_municipio_regiao_ano = ranking_municipio_regiao_ano.sort_values(by=criterio, ascending=False).head(10)

        # Use PrettyTable para criar a tabela formatada
        table = PrettyTable()
        table.field_names = ['Município', criterio]
        for _, row in ranking_municipio_regiao_ano.iterrows():
            table.add_row([row['Município'], row[criterio]])

        print(table)

        plt.figure(figsize=(10, 6))
        plt.barh(ranking_municipio_regiao_ano['Município'], ranking_municipio_regiao_ano[criterio])
        plt.xlabel(criterio)
        plt.ylabel('Município')
        plt.title(f'Top 10 Municípios em {regiao} - {ano}')
        plt.gca().invert_yaxis()
        plt.show()
        print()

# Exemplo:
criterio_usuario = 'Docentes'  
atualizar_tabela_e_grafico(criterio_usuario)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prettytable import PrettyTable

df = pd.read_csv("../raw_data/all_expanded_ML.csv")

def mostrar_tabela_e_grafico(ano, regiao, criterio):
    df_filtrado = df[(df['Ano'] == ano) & (df['Região'] == regiao)]

    df_filtrado = df_filtrado.sort_values(by=criterio, ascending=False).head(10)

    #tabela
    table = PrettyTable()
    table.field_names = ['Município', criterio]
    for _, row in df_filtrado.iterrows():
        table.add_row([row['Município'], row[criterio]])

    print(table)

    # Criar o gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.barh(df_filtrado['Município'], df_filtrado[criterio])
    plt.xlabel(criterio)
    plt.ylabel('Município')
    plt.title(f'Top 10 Municípios em {regiao} - {ano}')
    plt.gca().invert_yaxis()
    plt.show()

#Dropbox para seleção de ano, região e critério
ano_escolhido = int(input("Ano"))
regiao_escolhida = input("Região")
criterio_usuario = input("Critério")

mostrar_tabela_e_grafico(ano_escolhido, regiao_escolhida, criterio_usuario)
```
